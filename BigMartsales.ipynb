```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('train_v9rqX0R.csv')
```

```python
dataset.head()
```

```python
dataset['Item_Type'].unique()
```

```python
dataset['Outlet_Size'].fillna(dataset['Outlet_Size'].mode()[0],inplace=True)
```

```python
dataset['Item_Weight'].fillna(dataset['Item_Weight'].median(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
dataset['Item_Type_Combined'] = dataset['Item_Identifier'].apply(lambda x: x[0:2])
```

```python
dataset['Item_Type_Combined'] = dataset['Item_Type_Combined'].map({'FD':'Food',
                                                             'NC':'Non-Consumable',
                                                             'DR':'Drinks'})
```

```python
dataset.drop(['Item_Identifier','Outlet_Identifier','Item_Type'],axis=1,inplace=True)
```

```python
dataset.head()
```

```python
itemfatcontent=pd.get_dummies(dataset['Item_Fat_Content'],drop_first=True)
outletsize=pd.get_dummies(dataset['Outlet_Size'],drop_first=True)
outletlocation=pd.get_dummies(dataset['Outlet_Location_Type'],drop_first=True)
outlettype=pd.get_dummies(dataset['Outlet_Type'],drop_first=True)
itemtypecombined=pd.get_dummies(dataset['Item_Type_Combined'],drop_first=True)
```

```python
dataset=pd.concat([dataset,itemfatcontent,outletsize,outletlocation,outlettype,itemtypecombined],axis=1)
```

```python
dataset.drop(['Item_Fat_Content','Outlet_Size','Outlet_Location_Type','Outlet_Type','Item_Type_Combined'],axis=1,inplace=True)
```

```python
dataset.head()
```

```python
X=dataset.drop(['Item_Outlet_Sales'],axis=1)
y=dataset['Item_Outlet_Sales']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
#Randomized Search CV and #Hypeparameters

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
rf=RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X,y)
```

```python
rf_random.best_params_
```
